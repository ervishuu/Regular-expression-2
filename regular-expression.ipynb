```python
ps = PorterStemmer()
corpus = []

for i in range(len(train['text'])):
    rev = re.sub("[^a-zA-Z0-9]",' ',str(train['text'][i]))
    rev =  rev.lower()
    rev = rev.split()
    rev = [ps.stem(word) for word in rev if word not in set(stopwords.words('english'))]
    rev = ' '.join(rev)
    corpus.append(rev)
  
```

```python
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()

def cleanText(text):
    rev = re.sub("[^a-zA-Z0-9]",' ',text)
    rev =  rev.lower()
    rev = rev.split()
    rev = [lemmatizer.lemmatize(word) for word in rev if word not in set(stopwords.words('english'))]
    rev = ' '.join(rev)  
    return rev

train['cleanText'] = train['text'].apply(lambda x: cleanText(x))

train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
model = Pipeline(
    [('vect' ,CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', RandomForestClassifier())]

)
print(model)
```
